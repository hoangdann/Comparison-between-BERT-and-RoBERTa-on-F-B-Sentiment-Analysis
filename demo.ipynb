```python
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer, BertForSequenceClassification, Trainer, TrainingArguments, AutoModelForSequenceClassification
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
import torch
from transformers import (RobertaTokenizer,
                          RobertaForSequenceClassification,
                          Trainer,
                          TrainingArguments)
from keras.models import load_model
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from sklearn.model_selection import train_test_split
import pickle as pkl
from sklearn.preprocessing import LabelEncoder
import gradio as gr
import numpy as np
```

```python
tokenizer_bert = AutoTokenizer.from_pretrained('bert-base-uncased')
bert = "/content/drive/MyDrive/ICT- NQH/End Term/BERT/trained_model"

# model = BertForSequenceClassification.from_pretrained(path)
model_bert = AutoModelForSequenceClassification.from_pretrained(bert)
```

```python
roberta = '/content/drive/MyDrive/ICT- NQH/End Term/RoBERTa/model'
tokenizer_ro = RobertaTokenizer.from_pretrained('roberta-base')
model_ro = AutoModelForSequenceClassification.from_pretrained(roberta)
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re

# Move models to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_ro.to(device)
model_bert.to(device)



def predict_sentiment(text):


    roberta_inputs = tokenizer_ro(text, return_tensors="pt", padding=True).to(device)
    bert_inputs = tokenizer_bert(text, return_tensors="pt", padding=True).to(device)

    with torch.no_grad():
        # Get logits and probabilities for Roberta
        roberta_outputs = model_ro(**roberta_inputs)
        roberta_logits = roberta_outputs.logits
        roberta_probabilities = torch.nn.functional.softmax(roberta_logits, dim=-1)
        roberta_positive_prob = round(roberta_probabilities.squeeze(0)[1].item(),3)
        roberta_negative_prob = round(roberta_probabilities.squeeze(0)[0].item(),3)
        roberta_predicted_label = "Positive Review" if roberta_positive_prob > roberta_negative_prob else "Negative Review"

        # Get logits and probabilities for BERT
        bert_outputs = model_bert(**bert_inputs)
        bert_logits = bert_outputs.logits
        bert_probabilities = torch.nn.functional.softmax(bert_logits, dim=-1)
        bert_positive_prob = round(bert_probabilities.squeeze(0)[1].item(),3)
        bert_negative_prob = round(bert_probabilities.squeeze(0)[0].item(),3)
        bert_predicted_label = "Positive Review" if bert_positive_prob > bert_negative_prob else "Negative Review"

    return (
        roberta_predicted_label,
        roberta_positive_prob,
        roberta_negative_prob,
        bert_predicted_label,
        bert_positive_prob,
        bert_negative_prob,
    )
custom_theme = {
    "text": "#333333",  # Text color
    "background": "#f5f5f5",  # Background color
    "font_family": "sans-serif",  # Font family
    "heading1_color": "#007bff",  # Title color
    "heading2_color": "#4285f4",  # Description color
    "input_widget": {
        "background": "#e9ecef",  # Input textbox background
        "border_color": "#ced4da",  # Input textbox border color
        "text_color": "#333333",  # Input text color
    },
    "output_widget": {
        "background": "#e9ecef",  # Output textbox background
        "border_color": "#ced4da",  # Output textbox border color
        "text_color": "#333333",  # Output text color
    },
}

iface = gr.Interface(
    fn=predict_sentiment,
    inputs=gr.Textbox(label="Restaurant Review", placeholder="Write your review here…",lines=5),
    outputs=[
        gr.Textbox(label="Roberta Sentiment"),
        gr.Textbox(label="Roberta Positive Prob.",text_align="right"),
        gr.Textbox(label="Roberta Negative Prob.",text_align="right"),
        gr.Textbox(label="BERT Sentiment"),
        gr.Textbox(label="BERT Positive Prob.",text_align="right"),
        gr.Textbox(label="BERT Negative Prob.",text_align="right"),
    ],
    title="Restaurant Review Sentiment Analysis",
    description="Get sentiment predictions (positive or negative) along with probabilities from both Roberta and BERT models based on your food/restaurant review.",
    theme=custom_theme,
)

# Launch the Gradio interface
iface.launch()
```
